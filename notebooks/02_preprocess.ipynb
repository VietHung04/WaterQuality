TIỀN XỬ LÝ DỮ LIỆU

1. COPY DATA TỪ EDA SANG

```python
import pandas as pd

df = pd.read_csv('../data/raw/water_potability.csv')
df.head()
```

2. XỬ LÝ MISSING

```python
df.isnull().sum()
```

Dữ liệu có missing → xử lý bằng median

Không drop vì sẽ mất dữ liệu

```python
df.fillna(df.median(), inplace=True)
```

3. XỬ LÝ OUTLIER

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

Loại bỏ outlier giúp model chính xác hơn

4. CHUẨN HÓA DỮ LIỆU

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = df.drop('Potability', axis=1)
y = df['Potability']

X_scaled = scaler.fit_transform(X)
```

```python
import pandas as pd

X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
df_processed = X_scaled.copy()
df_processed['Potability'] = y.values
```

5. LƯU DATA SAU XỬ LÝ

```python
df_processed.to_csv('../data/processed/data_cleaned.csv', index=False)
```

Đã xử lý missing bằng median

Đã loại outlier bằng IQR

Đã chuẩn hóa dữ liệu
