KHÁM PHÁ DỮ LIỆU

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/raw/water_potability.csv')

# xem info
print(df.info())

# thống kê
print(df.describe())

# check missing
print(df.isnull().sum())
```

VẼ BIỂU ĐỒ

```python
# histogram
df.hist(figsize=(12,10))
plt.show()

# correlation
sns.heatmap(df.corr(), annot=True)
plt.show()

# boxplot
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```
