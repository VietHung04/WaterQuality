KHÁM PHÁ DỮ LIỆU

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/raw/water_potability.csv')

# xem info
print(df.info())

# thống kê
print(df.describe())

# check missing
print(df.isnull().sum())
```

* Dữ liệu gốc có 3276 dòng nhưng các cột quan trọng lại bị thiếu hụt dữ liệu: ph (chỉ có 2785 non-null), Sulfate (2495 non-null), và Trihalomethanes (3114 non-null).

VẼ BIỂU ĐỒ

```python
# histogram
df.hist(figsize=(12,10))
plt.show()

# correlation
sns.heatmap(df.corr(), annot=True)
plt.show()

# boxplot
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

* Hầu hết các đặc trưng (như ph, Hardness, Chloramines, Sulfate, Conductivity, Organic_carbon, Trihalomethanes, Turbidity) đều có phân phối gần giống với phân phối chuẩn (Normal Distribution), với đỉnh tập trung ở giữa và thon dần về hai phía.

* Riêng đặc trưng Solids (Tổng chất rắn hòa tan) có xu hướng lệch phải (right-skewed) rõ rệt.

* Biến mục tiêu Potability bị mất cân bằng. Cột mốc 0 (Nước không an toàn) xuất hiện với tần suất cao hơn hẳn so với cột mốc 1 (Nước an toàn).
